### 1 OpenCV

```python
import cv2

img = cv2.imread('Images/flower.jpg', cv2.IMREAD_ANYCOLOR)
resized_img = cv2.resize(img, (500,500))  # resize the image for smaller window

cv2.imshow('Flower', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2 Matplotlib

```python
import matplotlib.pyplot as plt

img = plt.imread('Images/flower.jpg')

plt.imshow(img)
plt.axis('off')
plt.show()
```

### 3 Pillow

```python
from PIL import Image

img = Image.open('Images/flower.jpg')
display(img)
```

### 4 Scikit-Image

```python
from skimage import io

img = io.imread('Images/flower.jpg')
io.imshow(img)

plt.axis('off')
plt.show()
```

### TensorFlow

```python
import tensorflow as tf
from tensorflow import io
from tensorflow import image

tf_img = io.read_file('Images/flower.jpg')
img = image.decode_png(tf_img, channels=3)

plt.imshow(img)
plt.axis('off')
plt.show()
```
